# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
basekeys=Base.classes.keys()
basekeys
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
inspector = inspect(engine)
```

```python
# View contents of 'measurement' and 'station' tables
for table in basekeys:
    print(f'Columns for {table} table:')
    columns=inspector.get_columns(table)
    for column in columns:
        print(column["name"], column['type'])
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

# Calculate the date 1 year ago from today
## Data stops at 2017-08-23. Query from last year returns an empty set 
## lastyear=dt.date.today()-dt.timedelta(days=365)
lastyear='2016-08-23'

# Perform a query to retrieve the data and precipitation scores
# yeardata=session.query(Measurement.date,Measurement.prcp).\
#     filter(Measurement.date > lastyear).\
#     group_by(Measurement.date).all()

# Query with 'None' data removed
yeardata=session.query(Measurement.date,func.avg(Measurement.prcp)).\
    filter(Measurement.date > lastyear).\
    filter(Measurement.prcp != None).\
    group_by(Measurement.date).all()

# Save the query results as a Pandas DataFrame and set the index to the date column
yeardata_df=pd.DataFrame(yeardata,columns=['Date','Precipitation'])
yeardata_df.set_index('Date',inplace=True)
yeardata_df
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
yeardata_df.plot()
ticklocs=[8,38,69,99,130,161,189,220,250,281,311,342]
ticklabels=['2016-09-01','2016-10-01','2016-11-01','2016-12-01','2017-01-01',
            '2017-02-01','2017-03-01','2017-04-01','2017-05-01','2017-06-01','2017-07-01','2017-08-01']
plt.xticks(ticklocs,ticklabels,rotation=-70,fontsize=10)
plt.title('Avg Rainfall Last Year')
plt.ylabel('Inches',fontsize=10)
plt.tight_layout()
plt.savefig('figures/precipitation.png')
```

## Use Pandas to calcualte the summary statistics for the precipitation data


```python
# How many stations are available in this dataset?
numstation=len(session.query(Station.station).distinct().all())
print(f"There are {numstation} stations available.")
```

```python
# What are the most active stations?
# List the stations and the counts in descending order.
activity=session.query(Measurement.station,func.count(Measurement.station)).\
    group_by(Measurement.station).\
    filter(Measurement.tobs != None).\
    order_by(func.count(Measurement.station).desc()).all()
activity
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?
mostactive=activity[0][0]
mostactivetemp=session.query(func.max(Measurement.tobs),
                        func.min(Measurement.tobs),
                        func.avg(Measurement.tobs)).\
    filter(Measurement.station==mostactive).\
    group_by(Measurement.station).all()
mostactivetemp
print(f'Temperatures from the most active station ({mostactive})')
print(f'High: {mostactivetemp[0][0]}')
print(f'Low: {mostactivetemp[0][1]}')
print(f'Avg: {mostactivetemp[0][2]}')
```

```python
# Choose the station with the highest number of temperature observations.
activity_lastyear=session.query(Measurement.station,func.count(Measurement.station)).\
    group_by(Measurement.station).\
    filter(Measurement.tobs != None).\
    filter(Measurement.date > lastyear).\
    order_by(func.count(Measurement.station).desc()).all()
mostactive_lastyear=activity_lastyear[0][0]
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
yeartemps=session.query(Measurement.date,Measurement.tobs).\
    filter(Measurement.station==mostactive_lastyear).\
    filter(Measurement.tobs != None).\
    filter(Measurement.date > lastyear).all()
yeartemps_df=pd.DataFrame(yeartemps,columns=['Date','Temp'])
yeartemps_df.set_index('Date',inplace=True)
yeartemps_df.hist(bins=12)
plt.xlabel('Temperature (F)',fontsize=10)
plt.ylabel('Samples',fontsize=10)
plt.title(f'{mostactive_lastyear} Temperatures Last Year')
plt.tight_layout()
plt.savefig('figures/station_histogram.png')
```

```python
# Write a function called `calc_temps` that will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
trip_start='2017-06-09'
trip_end='2017-06-18'
trip_temps=calc_temps(trip_start,trip_end)[0]
print(trip_temps)
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)
avg_temp=trip_temps[1]
ptp=trip_temps[2]-trip_temps[0]
avg_temp
```

```python
# Plot the results from your previous query as a bar chart. 
%matplotlib inline
plt.bar(0,avg_temp,yerr=ptp)
plt.xticks([])
plt.title('Trip Avg Temp')
plt.ylabel('Temperature (F)')
plt.savefig('figures/temperature.png')
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```

```python
# calculate the daily normals for your trip
# push each tuple of calculations into a list called `normals`

# Set the start and end date of the trip
trip_start='2017-06-09'
trip_end='2017-06-18'

# Use the start and end date to create a range of dates
trip_dates=[]
trip_date=trip_start
while trip_date <= trip_end:
    trip_dates.append(trip_date[5:]) #remove year from date string
    nextday=dt.datetime.strptime(trip_date, '%Y-%m-%d')+dt.timedelta(days=1)
    trip_date = nextday.strftime('%Y-%m-%d')

# Loop through the list of %m-%d strings and calculate the normals for each date
normals=[]
for trip_date in trip_dates:
    date_normals = daily_normals(trip_date)
    normals.append(date_normals[0])
    
normals
```

```python
# Load the previous query results into a Pandas DataFrame and add the `trip_dates` range as the `date` index
tmins=[normal[0] for normal in normals]
tavgs=[normal[1] for normal in normals]
tmaxs=[normal[2] for normal in normals]
trip_normals_df=pd.DataFrame({'date':trip_dates,'tmin':tmins,'tavg':tavgs,'tmax':tmaxs})
trip_normals_df
```

```python
# Plot the daily normals as an area plot with `stacked=False`
trip_normals_df.plot.area(xticks=range(10),ylim=(60,90),stacked=False)
plt.xticks(range(10),trip_dates,rotation=-70)
plt.title('Daily Normals for Trip Dates')
plt.ylabel('Temperature (F)',fontsize=10)
plt.xlabel('Dates',fontsize=10)
plt.tight_layout()
plt.savefig('figures/daily_normals.png')
```
